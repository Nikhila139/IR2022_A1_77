```python
#importing all the necessary libraries such as nltk, pandas, numpy,...
import nltk
import pandas as pd
import numpy as np
import os
import string
import io
import glob
import re
import json
import pickle
```

```python
#importing tokenizer, stopwords, stemmer and lemmatizer from nltk library
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords

from nltk.stem import PorterStemmer
ps = PorterStemmer()

from nltk.stem import WordNetLemmatizer
lem = WordNetLemmatizer()
```

```python
path="data_A1"
os.chdir(path)
```

```python
#Initializing doccuments list and posting list
doc={}
pl={}
```

```python
#pre_process function that converts to lower case , removes punctuations and also does tokenization and lemmatization
l=len(string.punctuation)
def pre_process(s):
    s = s.lower()
    s = s.translate(s.maketrans(string.punctuation,' '*l,''))
    s = re.sub('[^A-Za-z\s\n ]+', ' ',s)
    
    t = word_tokenize(s)
    t = [lem.lemmatize(w) for w in t if w not in stopwords.words('english') and w.isalpha()]
    return t
```

```python
#function used to build the dictionary
def build_dic(c,tl):
    d={}
    for t in tl:
        if t not in d:
            d[t]=c
    return d
```

```python
#function to generate posting list of terms
def merge_pl(d):
    for t in d:
        if t in pl:
            pl[t].append(d[t])
        else:
            pl[t]=[d[t]]
```

```python
#processing and reading the files with encoding
c=0
for file in glob.glob("*"):
    c+=1
    print(file)
    with open(file, 'r', encoding="utf8", errors="ignore") as f:
        text = f.read()
        #print(f)
        doc[c]=file
        
        tl=pre_process(text)
        d=build_dic(c,tl)
        merge_pl(d)
```

```python
#function used to perform x and y and it takes posting lists as it's arguments
def xandy(x,y):
    l=[]
    #x=pl[x]
    #y=pl[y]
    nx=len(x)
    ny=len(y)
    i=0
    j=0
    c=0
    while i<nx and j<ny:
        if x[i]==y[j]:
            l.append(x[i])
            i+=1
            j+=1
        elif x[i]<y[j]:
            i+=1
        else:
            j+=1
        c+=1
    return c,l
```

```python
#function used to perform x or y and it takes posting lists as it's arguments
def xory(x,y):
    l=[]
    #x=pl[x]
    #y=pl[y]
    nx=len(x)
    ny=len(y)
    i=0
    j=0
    c=0
    while i<nx and j<ny:
        if x[i]==y[j]:
            l.append(x[i])
            i+=1
            j+=1
        elif x[i]<y[j]:
            l.append(x[i])
            i+=1
        else:
            l.append(y[j])
            j+=1
            
        c+=1
        
    while i<nx:
        l.append(x[i])
        i+=1
    while j<ny:
        l.append(y[j])
        j+=1
    
    return c,l
```

```python
#function used to negation of x and it takes posting list as it's argument
def notx(x):
    l=[]
    for d in range(1,1134):
        if d not in x:
            l.append(d)
    return l
```

```python
#function used to perform x and not y and it takes posting lists as it's arguments
def xandnoty(x,y):
    l=[]
    #x=pl[x]
    #y=pl[y]
    noty=notx(y)
    c,l=xandy(x,noty)
    return c,l
```

```python
#function used to perform x or not y and it takes posting lists as it's arguments
def xornoty(x,y):
    l=[]
    #x=pl[x]
    #y=pl[y]
    noty=notx(y)
    c,l=xory(x,noty)
    return c,l
```

```python
#this function is used to process the query given by the user and call appropriate functions
def process_query(s,o):
    l=[]
    c=0
    doc={}
    #pre-processing the query by user
    t=pre_process(s)
    print(t)
    comp=0
    res=pl[t.pop(0)]
    
    for i in range(len(o)):
        t1=pl[t.pop(0)]
        op=o.pop(0)
        #else-if to call appropraite functions
        if op=="AND":
            c,l=xandy(res,t1)
        elif op=="OR":
            c,l=xory(res,t1)
        elif op=="AND NOT":
            c,l=xandnoty(res,t1)
        elif op=="OR NOT":
            c,l=xornoty(res,t1)
            
        comp+=c
        res[:]=l[:]
    
    return comp,res
```

```python
#take number of queries from user
print("number of queries: ")
n=int(input())
for _ in range(n):
    #take the input query from user
    print("Input query: ")
    s=input()
    #take sequence of operations to be performed on pre processed query seperated by commas
    print("Input operation sequence: ")
    o=list(input().split(','))
    for i in range(len(o)):
        o[i].strip()
        o[i].upper()
    #pre process the query
    comp,d=process_query(s,o)
    res_doc=[]
    for i in d:
        res_doc.append(doc[i])
    #printing the results such as number of documents matched and number of comparisions made
    print(f"Number of documents matched: {len(d)}")
    print(f"No. of comparisons required: {comp}")
    print(f"List of documents matched: ")
    print(res_doc)
```
